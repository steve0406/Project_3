```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sma
import sqlite3
```

```python
conn = sqlite3.connect("sqlite:///../input/COVID_Cases_Mortality_Features_By_Country.sqlite")


coutries_tourism_df    = pd.read_sql_query(' SELECT Country , Tourism  FROM COVID_data ORDER BY tourism DESC LIMIT 10', conn)

coutries_population_df = pd.read_sql_query(' SELECT Country , Population_Size  FROM COVID_data ORDER BY tourism DESC LIMIT 10', conn)

meanage_death_df       = pd.read_sql_query(' SELECT Mean_Age , Deaths  FROM COVID_data Group BY Mean_age ORDER BY Mean_age DESC LIMIT 10 ', conn)

new_df                 =pd.read_sql_query('select Country,Deaths,Cases,Population_Size,Mean_Age,Outdoor_Pollution_Death_Rate,Case_Rate,Case_Fatality_Ratio from COVID_data Group BY Country ORDER BY Deaths DESC LIMIT 10',conn)

Countries_death_df     = pd.read_sql_query(' SELECT Country , Deaths  FROM COVID_data Group BY Country ORDER BY Deaths DESC LIMIT 10 ', conn)

coutries_tourism_df.head()
coutries_population_df.head()
meanage_death_df.head()
new_df.head()
```

```python
coutries_tourism_df.head()
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i],ha = 'center',rotation='vertical')

x=coutries_tourism_df['Country']
y=coutries_tourism_df['Tourism']

plt.bar(x, y)

addlabels(x, y)
plt.title("Countries Vs Tourism")
plt.xlabel("Countries")
plt.align='center'
plt.ylabel("Tourism")
plt.xticks(rotation='vertical')
plt.yticks(rotation='vertical')
plt.figure(figsize=(100,20), dpi=80) 
plt.show()
```
